# Introdução ao pandas e medidas de dispersão

```python
import pandas as pd 
dict_medidas = {'idade': [15,18,25, 25,40,55,58,60,80], 
                'altura': [160,162,165,168,172,174,174, 174, 176]}
```

```python
df = pd.DataFrame.from_dict(dict_medidas)
df
```

# Medidas de posição

```python
#media 
df['idade'].mean()
df['altura'].mean()
#mediana
df['idade'].median()
df['altura'].median()
#moda
df['idade'].mode()
df['altura'].mode()
```

### Medidas de Dispersã0

#### Variancia é a distancia entre a media de cada um dos pontos elvado ao quadrado
##### ex: (15 - media)^2


```python
#variancia 
df.idade.var()
```

#### Desvio padrao é a variancia ^2, ela esta na medida dos dados nesse caso(ANOS)

```python
#desvio padrao
df.idade.std() #nessa medida diz o desvio 
```

#### Coeficiente de variação - IDADE
##### Desvio padrao / media

```python
#coediciente de variacao - idade

CVariacao = df.idade.std() / df.idade.mean() * 100
CVariacao #nessa medida avalia quantos % se varia dentre as amostras 
```

# Medidas de Forma Assimetria e Curtose


```python
#fazendo teste de assimetria para idade
df.idade.skew()
#valor positivo, logo a calda direitab do grafico eh mais longa, ou seja os valores do final estao mais separados
```

```python
#fazendo teste de assimetria para altura
df.altura.skew()
#contrario da idade, calda positiva eh maior
```

```python
#Curtose - Shape do Grafico -  idade 
df.idade.kurtosis() #lepcurtica, mais separados, logo mais achatada
```

```python
#Curtose - Shape do Grafico -  altura
df.altura.kurtosis() #lepcurtica, mais separados, logo mais achatada
```

```python
df.idade.describe() #metodo que rorna valores que precisaremos 
```

```python
#Correlação do DATAFRAME 
df.corr(method='pearson') #para apenas com duas variaveis
df.corr(method='spearman')
```

```python
#Correlação do DATAFRAME 
df.idade.corr(df.altura, method='spearman') #para duas variaveis especificas, se tiver mais que 2
```
