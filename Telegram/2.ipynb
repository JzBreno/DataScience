# Correlação

```python
import pandas as pd

TimeService = {'Meses' : [6,10,12,18,24,30,36,50]
                ,'Salario': [1800, 2000,2400,3000,3600,4300,5100,6000]}

df = pd.DataFrame.from_dict(TimeService)
df.corr(method='pearson') #correlação das variaveis  #pode usar pearson ou spearman em method
```

# Representações Graficas


```python
import matplotlib.pyplot as plt
plt.bar(df.Meses, df.Salario)
```

```python
#histogramas
df.Meses.hist()
```

### Criar DataFrame de Vendas

```python
import pandas as pd
#criando dicionario
dict_vendas = {'categoria' : ['masculino', 'feminino', 'infantil', 'casa'],
              'valor': [400000, 600000, 250000, 580000],
              'quantidade': [3000, 5000, 1500, 2500]}

#criando datframe apartir de dicionario
df_vendas = pd.DataFrame.from_dict(dict_vendas)
```

```python
#mostrando valores especificos
df_vendas.describe()
```

```python
df_vendas
```

```python
#Grafico de Barras Vertical - Valor

df_vendas.plot.bar(x='categoria', y='valor')
```

```python
#Grafico de Barras Vertical - Valor ordenado
#ascending diz se quer do maior para o menor ou nao, True or FAlse
df_vendas.sort_values('valor', ascending=False ).plot.bar(x='categoria', y='valor')
```

```python
#Garfico de barras horizontal - Quantidade
#ascending diz se quer do maior para o menor ou nao, True or FAlse
df_vendas.sort_values('quantidade', ascending=True).plot.barh(x='categoria', y='quantidade')
```

### Grafico de Dispersão 


```python

#dict de medidas
dict_medidas = {'idade': [15,18,25, 25,40,55,58,60,80], 
                'altura': [160,162,165,168,172,174,174, 174, 176]}

df_medidas = pd.DataFrame.from_dict(dict_medidas)
df_medidas
```

```python
#grafico de dispersão
df_medidas.plot.scatter(x='idade', y='altura')
```

```python
#r de pearson mostra que existe uma correlação forte positiva e que enqunato x aumenta, y aumenta 
df_medidas.corr(method='pearson')
```

```python
#Garfico de BloxPlot (Diagrama de Caixa ) - Idade

df_medidas.idade.plot.box()
```

```python
#grafico de Bloxplot - ALtura
df_medidas.altura.plot.box()
```

```python
#Gerar Dataframe com dads temporais

dict_faturamento = {'data_ref': ['2019-01-01','2020-01-01','2021-01-01', '2022-01-01', '2023-01-01'],
                    'valor': [400000, 800000, 500000, 800000, 900000]}

df_faturamento = pd.DataFrame.from_dict(dict_faturamento)
df_faturamento
```

```python
#conversao de tipos de variaveis
df_faturamento.info()
```

```python
#tranformando data_ref de datetime para datetime
df_faturamento['data_ref'] = pd.to_datetime(df_faturamento['data_ref']) 
df_faturamento.info()
```

```python
#gerando grafico de linhas
df_faturamento.plot.line(x='data_ref', y='valor')
```
